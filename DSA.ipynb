```python
nums = [5, 1, 101, 99, 2, 13, 19, 21, -1, 89]

# lets create 2 pointers p1 and p2


def max_in_list(nums):
  p1 = 0
  max = 0
  while p1 < (len(nums)):
    if nums[p1] > max:
      max = nums[p1]
    p1 += 1

  return max

max_in_list(nums)
```

```python
# nums = [5, 1, 101, 99, 2, 13, 19, 21, -1, 89]
# nums = [1,2,3,4,5]
# nums = [-1,-2,-3]
nums = [-3,-2,-1,0,1,2,3]

def is_sorted(nums):
  p1 = 0
  p2 = 1
  while p2 < len(nums):
    if nums[p1] < nums[p2]:
      p1 += 1
      p2 += 1
    else:
      return False
  return True

is_sorted(nums)
```

```python
nums = [5, 3, 7, 3, 10, 14, 15, 100, -1, -2]
target = -1

def index_of(nums, target):
  p1 = 0
  while p1 < len(nums):
    if nums[p1] != target:
      p1 += 1
    elif nums[p1] == target:
      return p1
    else:
      return -1


index_of(nums, target)
```

```python
nums = [1, 4, 5, 2, 3]


def reverse_list(nums):
  p1 = 0
  rev = []
  while len(nums) != 0:
    rev.append(nums[-1])
    nums.pop()
  return rev

reverse_list(nums)
```

```python
# a = [1, 3, 5], b = [2, 4, 6] → [1, 2, 3, 4, 5, 6]

a = [1, 3, 5]
b = [2, 4, 6]

def merge_sorted(a, b):
  p1 = 0
  p2 = 0
  p3 = 0
  lst = []
  while p1 < len(a):
    if len(lst) != 0:
      if a[p1] < b[p2] and lst[p3] < a[p1]:
        lst.append(a[p1])
        p1 += 1
        p3 += 1
      elif b[p2] < a[p1] and lst[p3] < b[p2]:
        lst.append(b[p2])
        p2 += 1
        p3 += 1
    else:
      if a[p1] < b[p2]:
        lst.append(a[p1])
      elif b[p2] < a[p1]:
        lst.append(b[p2])
  return lst



merge_sorted(a, b)
```

```python
# [1, 2, 3, 4] → [1, 3, 6, 10]
nums = [1, 2, 3, 4]

def prefix_sums(nums):
  p1 = 0
  p2 = 0
  sums = []
  while p1 < len(nums):
    if len(sums) == 0:
      sums.append(nums[p1])
    p1 += 1
    if len(sums) != len(nums):
      new_val = sums[p2] + nums[p1]
      sums.append(new_val)
      p2 += 1
  return sums

prefix_sums(nums)
```

```python
# [0, 1, 0, 3, 12] → [1, 3, 12, 0, 0]


nums = [0, 1, 0, 3, 12]

def move_zeros(nums):
  lst = []
  lst_zero = []
  p1 = 0
  while p1 < len(nums):
    if nums[p1] != 0:
      lst.append(nums[p1])
    else:
      lst_zero.append(nums[p1])
    p1 += 1
  lst.extend(lst_zero)
  return lst

move_zeros(nums)
```
